```python
result = requests.get('https://en.wikipedia.org/wiki/Maya_(M.I.A._album)')
```

```python
soup = bs4.BeautifulSoup(result.text, 'lxml')
```

```python
my_image = soup.select('.thumbimage')
```

```python
my_image
```

```python
image_to_download = my_image[0]
```

```python
image_to_download['src']
```

```python
image_link = requests.get('https:' + image_to_download['src'])
```

```python
#Binary content for image
image_content = image_link.content
```

```python
image_content
```

```python
#open the file to write in binary mode (wb)
f = open('downloaded_image1.jpg','wb')
f.write(image_content)
f.close()
```

```python
#www.toscrape.com
```

```python
book_title_rating_three = []
for current_page in range(1,51):
        
    page_result = requests.get(f'http://books.toscrape.com/catalogue/page-{current_page}.html')
    
    soup = bs4.BeautifulSoup(page_result.text, 'lxml')
    three_ratings_book_set = soup.select('.star-rating.Three')
    
    for books_with_three_rating in three_ratings_book_set:
        book_header_tags = books_with_three_rating.parent.find_next('h3').find('a')
        book_title_rating_three.append(book_header_tags['title'])
    
print('The books with three star ratings are:')
for book_name in book_title_rating_three:
    print(book_name)
```
